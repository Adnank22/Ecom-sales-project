```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Adnank22/Ecom-sales-project/main/Ecommerce%20Sales%20Report.csv')
```

```python
#drop unlrelated/blank comlumns
df.drop(['New','PendingS'],axis=1,inplace=True)
```

```python
df.info()
```

```python
#checking null value
df.isnull().sum()
```

```python
#drp null values
df.dropna(inplace=True)
```

```python
#change datatype
df['Date']=pd.to_datetime(df['Date'])
df['ship-postal-code']=df['ship-postal-code'].astype('int')
```

```python
#rename columns
df.rename(columns={'Qty':'Quantity'})
```

# **Exploatory Data Analysis**

```python
df.columns
```

**Size Column**

```python
ax= sns.countplot(x='Size',data=df)
```

```python
#adding data labels to the bar chart
ax= sns.countplot(x='Size',data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

**Observation: From above graph, "M" Size is the most sold in the size category**

## **Group By**

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
s_qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size', y='Qty', data=s_qty)
```

```python
#add colour in the chart
sns.barplot(x='Size', y='Qty', data=s_qty, palette='rocket')
```

**Courier Status**

```python
sns.countplot(data=df, x='Courier Status', hue="Status" )
```

**Observaton: From the above chart it shows that most of the orders are in "shipped" status**

```python
df["Size"].hist()
```

```python
df["Category"]= df["Category"].astype(str)
column_data=df["Category"]
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

**Observation: The above graph indicates that the best selling category is "T-shirt"**

```python
#Checking B2b data by using pie chart
B2B_check = df["B2B"].value_counts()

#plot the pie chart
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')
plt.show()
```

**Observation: From the above chart, we can see that the 99.2% buyers are retailers and only 2% are B2B Buyers**

```python
#prepare data for scatter plot
x_data = df["Category"]
y_data = df["Size"]

plt.scatter(x_data, y_data)
plt.xlabel("Category")
plt.ylabel("Size")
plt.title("Scatter Plot")
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(x='ship-state', data=df)
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=90)
plt.show()
```

```python
#distribution of only top 10 states
top_10_states = df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.countplot(x='ship-state', data=df, order=top_10_states.index)
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=45)
plt.show()
```

**Observation: Most of the buyers are from Maharashtra**
